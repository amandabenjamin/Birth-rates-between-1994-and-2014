# Data Visualization

What month has the most birthdays?

```python
#create a bar graph to display the number of births (numerical data) vs. the month (nominal/category)

import seaborn as sns #import seaborn libary
import matplotlib.pyplot as plt #import matplotlib libary

fig, ax = plt.subplots(1,1,figsize=(10,10)) #create a singular figure
ax = sns.barplot(x='month', y='births', data=df).set(xlabel = 'Months',
                                                    ylabel = 'Number (#) of Births',title = 'Number of Births per Month (1994 - 2014)');
```

From our bar chart, we see that September has the most birthdays. We also see an increasing trend in birthdays as the year progresses, peaks during September and then is seen to decrease untill the year ends.  

Do the number of birthdays increase from 1994 to 2014? 

```python
fig, ax = plt.subplots(1,1,figsize=(10,10)) #create a singular figure
sns.set(style="ticks")
ax = sns.lineplot(data=df, x='year', y='births',marker='o',
                  ci = None).set(xlabel = 'Year', ylabel = 'Number (#) of Births',title = 'Number of Births per Year (1994 - 2014)');
plt.grid()
```

Births were on the rise from 1994 untill 2006, when they hit their pinicle. After 2006, the line plot shows the number of births to decrease untill 2013, when they are seen to slighly increase again in the following year. 

How common are May birthdays? 

```python
#total births of the year
#total births during may

plt.subplots(1,1,figsize=(15,10)) #create a singular figure
ax1 = sns.barplot(x='year', y='births', data=df, color = 'blue',label='Full Year');

#make a new data set of just may birthdays
tmp = df[df['month'] == 5];

ax2 = sns.barplot(x='year', y='births', data=tmp,color = 'orange',label='May only').set(xlabel = 'Year',
                                                    ylabel = 'Number (#) of Births',title = 'How do May Birthdays Compare to Year Totals');
plt.legend();
```

May birthdays comprise a large number of birthdays throughout the year.
